```python
# TUGAS ANALISI DATA APBN INDONESIA 
# FENDI PRABOWO KELAS C

# VARIABEL YANG TERDAPAT PADA DATASET:
# 1. Tahun
# 2. Sumber_pendapatan
# 3. Keterangan_sumber_pendapatan
# 4. Nilai

import pandas as pd
```

```python
data_pendapatan = pd.read_csv('processed-apbn-pendapatan-indonesia-1990-2013.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_pendapatan
```

```python
# DISKRIPSI DATA 
# Sumber Pendapatan Pemerintah APBN

#Dataset ini berisi mengenai pos pendapatan Pemerintah di APBN

#Penjelasan mengenai Variabel pada Dataset ini:

#tahun: Tahun APBN
#sumber_pendapatan: Sumber Pendapatan Pemerintah (Pajak / Bukan Pajak / Hibah)
#keterangan_sumber_pendapatan: Keterangan dari Sumber Pendapatan tersebut
#nilai: Nilai dari Pos Pendapatan tersebut (dalam Milyar Rupiah)
#Catatan:

#Sampai dengan tahun 2004, realisasi menggunakan angka Perhitungan Anggaran Negara (PAN).
#Sejak tahun 2005, menggunakan angka Laporan Keuangan Pemerintah Pusat (LKPP).
#Sejak Januari 2012, data sudah disesuaikan dengan publikasi resmi KemenKeu.
# Sumber Data = https://data.go.id/dataset/sumber-pendapatan-pemerintah-apbn
```

```python
data_pendapatan.count()
```

```python
# STATISTIK TENTANG DATA 

data_pendapatan.columns #Untuk mengecek header
```

```python
s = data_pendapatan.nilai.sum()
mi = data_pendapatan.nilai.min()
ma = data_pendapatan.nilai.max()
me = data_pendapatan.nilai.mean()
md = data_pendapatan.nilai.median()
mo = data_pendapatan.nilai.mode()
miss = data_pendapatan.sumber_pendapatan.min()
mass = data_pendapatan.sumber_pendapatan.max()
mist = data_pendapatan.keterangan_sumber_pendapatan.min()
mast = data_pendapatan.keterangan_sumber_pendapatan.max()

print("Total pendapatan Pemerintah di APBN\t\t:",s)
print("Rata-rata pendapatan Pemerintah di APBN\t\t:",me)
print("Median data pendapatan Pemerintah di APBN\t:",md)
print("Pendapatan tertinggi Pemerintah di APBN berasal dari",mass, "sebesar", ma, "yang bersumber dari", mast)
print("Pendapatan terendah Pemerintah di APBN berasal dari",miss, "sebesar", mi, "yang bersumber dari", mist)
```

```python
data_pendapatan.groupby('sumber_pendapatan').sum()
```

```python
data_pendapatan.groupby('tahun').sum()
```

```python
data_pendapatan.groupby(['sumber_pendapatan','tahun'])['nilai'].sum()
```

```python
# Visualisasi Sederhana
ax = data_pendapatan.groupby('sumber_pendapatan').sum()['nilai']
ax = ax.plot(kind='pie', title ="Jumlah Pendapatan Pemerintah di APBN Berdasarkan Sumber Pendapatan ", legend=True, figsize=(10,10), fontsize=10)
```

```python
ax = data_pendapatan.groupby('tahun').sum()['nilai']
ax = ax.plot(kind='bar', title ="Jumlah Pendapatan Pemerintah di APBN per Tahun", legend=True, figsize=(15,15), fontsize=10)
ax.set_xlabel("Tahun", fontsize=10)
ax.set_ylabel("Nilai", fontsize=10)
```
